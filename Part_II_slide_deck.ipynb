## Dataset Overview

> The dataset used for the analysis is a loan dataset from Prosper which consists of 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.

```python
# import all necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('prosperLoanData.csv')
```

```python
# function to plot bar chart
def hist(Dataframe,variable, x_label, title):
    #Add figure size
    plt.figure(figsize = [12, 5])
    base_color = sb.color_palette()[2]
    sb.countplot(data=Dataframe, x=variable, color  = base_color );
    plt.xlabel(x_label)
    plt.title(title, fontsize=12)
    plt.show()
```

## Plot Distribution of ProsperScore
> Based on the graph below, it can be observed that most borrowers have ProsperScore between 4 and 8 and 4 having the highest count. ProsperScore is a categorical variable which takes values from 1 to 10.

```python
hist(df,"ProsperScore","ProsperScore","Distribution of ProsperScore")
```

## Plot Distribution of IncomeVerifiable

> It can be seen from the plot below that those whose sources of income are verifiable have greater chance of receiving loan. It is however a great determinant of prosperScore.

```python
hist(df,"IncomeVerifiable", "IncomeVerifiable","Plot Distribution of IncomeVerifiable")
```

## IsBorrowerHomeowner vs ProsperScore
It can be seen from the graph below that the ProsperScore of borrowers who are home owners is higher than those who are not.

```python
def bivariateBar(variable, hueVar):
    plt.figure(figsize = [12, 5])
    sb.countplot(data=df, x=variable, hue=hueVar);
    plt.show()
bivariateBar('ProsperScore','IsBorrowerHomeowner');
```
